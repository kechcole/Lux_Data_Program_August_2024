## 3. Answers to quiz. 

### 3.1 Find all records where the weather was exactly clear.

```python
# Fetch all the rows where weather is clear 

data.loc[data['Weather'] == 'Clear']
```

### 3.2 Find the number of times the wind speed was exactly 4 km/hr.

```python
# First select rows with windspeed of 4, save them into a variable 
speed_4KM = data.loc[data['Wind Speed_km/h'] == 4]

# Find the length of features/number of rows  in the above subset data
print(f'Therea are {len(speed_4KM)} features ')
```

### 3.3 Check if there are any NULL values present in the dataset.

```python
# Any column with null values, columns will be True if they have null, else False 

has_Null = data.isnull().any()
has_Null
```

### 3.4 Rename the column "Weather" to "Weather_Condition."

```python
# Rename and save , place values into a dictionary 

data.rename(columns={
    # Originale : New name 
    'Weather': 'Weather_Conditions'}, inplace=True)

data
```

### 3.5 What is the mean visibility of the dataset?

```python
# Determine the mean of visibility column 

print(f'The mean of visisbility rounded to 2 decimal places is : {round(data.Visibility_km.mean(), 2)}')
```

### 3.6 Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
# Get rows 
wd_SpeedAbove_24 = data[data['Wind Speed_km/h'] > 24]

# Find length of dataset 
print(f'There are {len(wd_SpeedAbove_24)} observations ')
```

### 3.8 What is the mean value of each column for each weather condition?

```python
# Sunset numeric colums into one list by exluding all objects  

numeric_Columns = data.select_dtypes(exclude=['object']).columns.to_list()

# Find mean for each 
data[numeric_Columns].mean()
```

### 3.9 Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
# Filter using logical conditions 

# Weather conditions is clear and humidity is above 50 
data[(data['Weather_Conditions'] == 'Clear') & (data['Rel Hum_%'] > 50)]
```

```python
# Weather conditions is clear and visibility is above 40 
data[(data['Weather_Conditions'] == 'Clear') & (data['Visibility_km'] > 40)]
```

### 3.10 Find the number of weather conditions that include snow.

```python
# Filter and save to variable 
snowy_Weather = data[data['Weather_Conditions'] == 'Snow']
print(f'Observations with snow : \n {snowy_Weather}\n')


print(f'Number of observations {len(snowy_Weather)}')
```
